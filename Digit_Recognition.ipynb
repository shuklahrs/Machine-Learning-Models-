#**DIGIT RECOGNITION USING RANDOM FOREST CLASSIFIER**

#Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Importing Dataset

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

#Segragating Dataset

```python
X = dataset.iloc[ : , 1: ]
print(X)
X.shape
```

```python
Y = dataset.iloc[:,0]
print(Y)
Y.shape
```

#Splitting dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
X_train.head()
```

```python
y_train.head()
```

#Training Model Using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

#Predicting the Test Set

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

#Model Accuracy

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score: {0}%".format(accuracy_score(y_test, y_pred) * 100))
```

#Visualising the Test Set

```python
index = 155

print("Prediction: " + str(classifier.predict(X_test)[index]))
plt.axis('off')
plt.imshow(X_test.iloc[index].values.reshape(28,28), cmap = 'gray')
```
